# Class Activity 4: Metropolis-Hastings reversibility [15 minutes]


The ***Metropolis-Hastings (MH)*** transition distribution $\theta^{(t)} \sim p( \theta^{(t)}|\theta^{(t-1)})$ is a ***Markov chain*** defined by the ***perturbation proposal distribution*** 

$$\tilde \theta^{(t)} \sim \tilde p(\tilde \theta^{(t)}|\theta^{(t-1)}) \quad \text{ with } \quad 
\theta^{(t)} = \left\{\begin{array}{ll}
\tilde \theta^{(t)} & \text{with probability } \min\left(1,\frac{p(\tilde \theta^{(t)}\color{gray}{|x})}{p(\theta^{(t-1)}\color{gray}{|x})}\frac{\tilde p(\theta^{(t-1)}|\tilde \theta^{(t)})}{\tilde p(\tilde \theta^{(t)}|\theta^{(t-1)})}\right)\\
\theta^{(t-1)} & \text{otherwise}
\end{array}\right. $$

A ***Markov chain*** is called ***reversible*** if 

$$p(\theta^{(t)}|\theta^{(t-1)})p(\theta^{(t-1)}\color{gray}{\color{gray}{|x}}) = p(\theta^{(t-1)}|\theta^{(t)})p(\theta^{(t)}\color{gray}{\color{gray}{|x}})$$

If ***MH chains*** are ***reversible*** then

$$\frac{p(\theta^{(t)}\color{gray}{|x})}{p(\theta^{(t-1)}\color{gray}{|x})} = \frac{p(\theta^{(t)}|\theta^{(t-1)})}{p(\theta^{(t-1)}|\theta^{(t)})} \quad \overset{\text{which for}}{\text{MH is}} \quad \frac{p(\theta^{(t)}\color{gray}{|x})}{p(\theta^{(t-1)}\color{gray}{|x})} = \frac{\min\left(1, \frac{p(\theta^{(t)}\color{gray}{|x})}{p(\theta^{(t-1)}\color{gray}{|x})} \frac{\tilde p(\theta^{(t-1)}|\theta^{(t)})}{\tilde p(\theta^{(t)}|\theta^{(t-1)})}\right)}{\min\left(1, \frac{p(\theta^{(t-1)}\color{gray}{|x})}{p(\theta^{(t)}\color{gray}{|x})}\frac{\tilde p(\theta^{(t)}|\theta^{(t-1)})}{\tilde p(\theta^{(t-1)}|\theta^{(t)})}\right) } \frac{\tilde p(\theta^{(t)}|\theta^{(t-1)})}{\tilde p(\theta^{(t-1)}|\theta^{(t)})}$$

# Metropolis-Hastings [8 minutes]


The ***Metropolis-Hastings*** transition distribution $\theta^{(t)} \sim p( \theta^{(t)}|\theta^{(t-1)})$ is defined by the proposal distribution 

$$\tilde \theta^{(t)} \sim \tilde p(\tilde \theta^{(t)}|\theta^{(t-1)}) \quad \text{ with } \quad 
\theta^{(t)} = \left\{\begin{array}{ll}
\tilde \theta^{(t)} & \text{with probability } \min\left(1,\frac{p(\tilde \theta^{(t)}|x)}{p(\theta^{(t-1)}|x)}\frac{\tilde p(\theta^{(t-1)}|\tilde \theta^{(t)})}{\tilde p(\tilde \theta^{(t)}|\theta^{(t-1)})}\right)\\
\theta^{(t-1)} & \text{otherwise}
\end{array}\right. $$

For ***symmetric proposal distributions*** $\frac{\tilde p(\theta^{(t-1)}|\tilde \theta^{(t)})}{\tilde p(\tilde \theta^{(t)}|\theta^{(t-1)})} = 1$, the ***acceptance probability*** reduces to
$\quad \min\left(1,\frac{p(\tilde \theta^{(t)}|x)}{p(\theta^{(t-1)}|x)}\right)$

Notice also that if the ***normalizing constant*** $c$ of the ***posterior distribution*** is not known the density ratio can still be calculated...

$$\frac{p(\tilde \theta^{(t)}|x)}{p(\theta^{(t-1)}|x)} = \frac{\cancel{c}g(\tilde \theta^{(t)}|x)}{\cancel{c}g(\theta^{(t-1)}|x)} \quad \text{Why is this interesting?} \quad \frac{p(\tilde \theta^{(t)}|x)}{p(\theta^{(t-1)}|x)} = \frac{\frac{1}{p(x)}p(\tilde \theta^{(t)},x)}{\frac{1}{p(x)}p(\theta^{(t-1)},x)} = \frac{\cancel{\frac{1}{p(x)}}p(\tilde \theta^{(t)},x)}{\cancel{\frac{1}{p(x)}}p(\theta^{(t-1)},x)}$$

The ***Metropolis-Hastings*** algorithm doesn't need the ***posterior distribution***... just the ***joint distribution***...

# Metropolis within Gibbs [5 minutes]

```python
from scipy import stats; import numpy as np; import matplotlib.pyplot as plt 
mu_0,sigma_0=0,10
loc_prior = stats.norm(loc=mu_0, scale=sigma_0) # mean=mu_0 / std=sigma_0
scale_prior = stats.halfnorm(scale=5) # std=1
df_prior = stats.randint(low=1, high=101) # df ~ Unif{1, 2, ..., 100}

n,loc_true,scale_true,df_true = 100,10,2,15 # True values data will be from... IRL, we don't know these values
x = stats.t(df=df_true, loc=loc_true, scale=scale_true).rvs(n) # Data from location-scale-t distribution

loglikelihood = lambda loc,scale,df: stats.t(loc=loc, scale=scale, df=df).logpdf(x).sum()
logjoint_eval_at_x = lambda loc,scale,df: (loglikelihood(loc,scale,df)+\
                                           loc_prior.logpdf(loc)+scale_prior.logpdf(scale)+df_prior.logpmf(df))

loc_perturbation_MH_proposal_tuning_parameter = 0.1
loc_perturbation_MH_proposal_distribution = stats.norm(scale=loc_perturbation_MH_proposal_tuning_parameter)
# `loc_prior`  : p(mu(t-1)|mu*)/p(mu*|mu(t-1)) = 1 since it's a symmetric proposal

# ... going to use `scale_prior` and `df_prior` as fixed i.i.d. MH proposal distribuitions
scale_fixed_iid_MH_proposal_distribution = scale_prior # tuning parameters taken to be prior hyperparameters
df_fixed_iid_MH_proposal_distribution = df_prior # tuning parameters taken to be prior hyperparameters
# `scale_prior`: p(sd(t-1)|sd*)/p(sd*|sd(t-1)) = p(sd(t-1))/p(sd*) is an easy pdf ratio
# `df_prior`   : p(df(t-1)|df*)/p(df*|df(t-1)) = (1/100)/(1/100) = 1 can be ignored
```

# Metropolis within Gibbs [6 minutes]

```python
T=1000; loc_MH_samples,scale_MH_samples,df_MH_samples = np.ones(T),np.ones(T),np.ones(T)
for t in range(1,T):
    
    # loc
    current_loc = loc_MH_samples[t-1]; proposed_loc = current_loc + loc_perturbation_MH_proposal_distribution.rvs()    
    log_acceptance_probability = logjoint_eval_at_x(proposed_loc, scale_MH_samples[t-1], df_MH_samples[t-1])-\
                                 logjoint_eval_at_x(current_loc, scale_MH_samples[t-1], df_MH_samples[t-1])
    loc_MH_samples[t] = [current_loc,proposed_loc][int(stats.uniform().rvs() < np.exp(log_acceptance_probability))]

    # scale
    current_scale = scale_MH_samples[t-1]; proposed_scale = scale_fixed_iid_MH_proposal_distribution.rvs()
    logtarget_ratio = logjoint_eval_at_x(loc_MH_samples[t], proposed_scale, df_MH_samples[t-1])-\
                      logjoint_eval_at_x(loc_MH_samples[t], current_scale, df_MH_samples[t-1])
    logMH_proposal_correction_factor = scale_fixed_iid_MH_proposal_distribution.logpdf(current_scale)-\
                                       scale_fixed_iid_MH_proposal_distribution.logpdf(proposed_scale)
    acceptance_probability = np.exp(logtarget_ratio+logMH_proposal_correction_factor)
    scale_MH_samples[t] = [current_scale,proposed_scale][int(stats.uniform().rvs() < acceptance_probability)]

    # df
    current_df = df_MH_samples[t-1]; proposed_df = df_fixed_iid_MH_proposal_distribution.rvs()
    log_acceptance_probability = logjoint_eval_at_x(loc_MH_samples[t], scale_MH_samples[t], proposed_df)-\
                                 logjoint_eval_at_x(loc_MH_samples[t], scale_MH_samples[t], current_df)
    df_MH_samples[t] = [current_df,proposed_df][int(stats.uniform().rvs() < np.exp(log_acceptance_probability))]
```

# Hamiltonian Dynamics [10 minutes]

A ***log joint probability distribution*** with parameters $\theta$  of interest and ***auxilliary variables*** $v$ with $\dim(v)=\dim(\theta)$ 

$$\Large
\begin{align*}
H(\theta, z) = {} & U(\theta) + K(v|\theta) \\
\text{often simplified to } \Longrightarrow \quad = {} & U(\theta) + K(v) \quad\quad\quad\quad\quad\quad\quad\quad\quad\quad\quad
\end{align*}$$

determines ***Hamiltonian dynamics*** of ***potential*** $U(\theta)$ and ***kinetic*** $K(v)$ ***energy*** (***momentum*** or ***velocity***)<br>which obey the ***law of conservation of energy*** defined by the differential equations 

$$\Large \begin{align*} 
\frac{d\theta}{dt} & = {} \frac{dH}{dv}(v(t)) & \frac{dv}{dt} & = {}  -\frac{dH}{d\theta}(\theta(t))\\
\end{align*}$$

Specifically, $\theta$ and $v$ ***energy*** parameters of a ***Hamiltonian dynamics system*** are evolved by differential equations across time $\theta_0 \overset{time}{\rightarrow} \theta_t$ and $v_0 \overset{time}{\rightarrow} v_t$ so the log joint density (***energy***) stays constant $p(\theta_0, v_0|x) = p(\theta_t, v_t|x)$ 

> The ***potential energy*** $U(\theta)$ and ***kinetic*** $K(v)$ ***energy*** remain inversely synchronized over time so that the total ***energy*** overall always remains constant

# Hamiltonian Monte Carlo (HMC) is just<br>Smart Metropolis-Hastings [7 minutes]

1. Augment the density of the $\theta$ parameter of interest with auxilliary parameter $v$<br>to produce a Hamiltonian dynmics system to produce smart proposals
    - Moving (theoretically) deterministically on probability distribution contours isn't random sampling
    - Countours aren't known in advance, so can't be sampled from ***uniformly*** (but step counts can be random)


2. Approximate ***Hamiltonian dynamics*** using ***differential equations*** producing ***parameter evolutions*** $\theta_0 \overset{time}{\rightarrow} \theta_t$ and $v_0 \overset{time}{\rightarrow} v_t$ with a random number of discrete steps $t$ and then perturb the ***momentum*** to produce the final proposal $(\theta_t, v_t+\epsilon_t)$ from initial value $(\theta_0, v_0)$ for a ***Metropolis-Hastings*** ***accept-reject step***

$$\min\left(1,\frac{p(\theta_t, v_t+\epsilon|x)}{p(\theta_0, v_0|x)}\frac{\tilde h(\theta_0, v_0|\theta_t, v_t+\epsilon)}{\tilde h(\theta_t, v_t+\epsilon|\theta_0, v_0)}\right)$$

# Symplectic Integration [5 minutes]


Consider the ***energy function*** $H(\theta, v)$ made from the ***potential energy*** $U(\theta)$ and ***kinetic energy*** $K(v)$ 

$$H(\theta, v) = U(\theta) + K(v)$$

which is proportional to a ***log of an independent joint distribution*** 

$$H(\theta, v) \propto \log p(\theta, v) = \log[q(\theta)h(v)], \theta \perp\!\!\!\perp v$$

such as an ***isotropic bivariate normal distribution***, in which case

$$ U(\theta) = \frac{1}{2}\theta^2 \quad \text{ and } \quad K(v) = \frac{1}{2}v^2$$

and ***conservation of energy*** $U(\theta) + K(v)$ in this context actually just turns out to imply that $\theta + v$ must be constant as well, since according to the ***Hamiltonian dynamics*** differential equations


\begin{align*} \frac{d\theta}{dt} & = {} \frac{dH}{dv} & \frac{dv}{dt} & = {}  -\frac{dH}{d\theta} \\ & = {} \frac{dK}{dv} = v & & = {}   -\frac{dU}{d\theta}= -\theta  \end{align*}

which ensures that changes in $\theta$ and $v$ over time $t$ are equal and opposite

# Symplectic Integration [8 minutes]


The following numerically approximate the solution trajectory of a differential equations.  These processes are referred to as ***integration*** because they add up all changes over time (like how ***integration*** add up areas under a curve)

1. ***Euler's method***: not a ***symplectic integrator***

| | | | | |
|-|-|-|-|-|
|| $ \begin{align*} 
\theta(t+\epsilon) & = {} \theta(t) + \epsilon \frac{d\theta(t)}{d t}\\
& = {} \theta(t) + \epsilon v(t)
\end{align*}$ | $\quad$ | $  \begin{align*} 
v(t+\epsilon) & = {} v(t) + \epsilon \frac{dv(t)}{d t}\\
& = {} v(t) - \epsilon \theta(t)
\end{align*}$ ||

2. ***Euler's method*** (***improved variant***): a ***first order symplectic integrator***

| | | | | |
|-|-|-|-|-|
|| $ \begin{align*} 
\theta(t+\epsilon) & = {} \theta(t) + \epsilon \frac{d\theta(t)}{d t}\\
& = {} \theta(t) + \epsilon v(t)
\end{align*}$ | $\quad$ | $  \begin{align*} 
v(t+\epsilon) & = {} v(t) + \epsilon \frac{dv(t+\epsilon)}{d t}\\
& = {} v(t) - \epsilon \theta(t+\epsilon)
\end{align*}$ ||

3. The ***leapfrog method*** (or ***leapfrog integration***): a  ***second order symplectic integrator***

| | | | | | |
|-|-|-|-|-|-|
|| $  \begin{align*} 
\theta(t+\epsilon/2) & = {} \theta(t) + \frac{\epsilon}{2} \frac{d\theta(t)}{d t}\\
& = {} \theta(t) + \frac{\epsilon}{2} v(t)
\end{align*}$ | $\quad$ | $  \begin{align*} 
v(t+\epsilon) & = {} v(t) + \epsilon \frac{dv(t+\epsilon/2)}{d t}\\
& = {} v(t) - \epsilon \theta(t+\epsilon/2)
\end{align*}$ | $\quad$ | $  \begin{align*} 
\theta(t+\epsilon) & = {} \theta(t + \epsilon/2) + \frac{\epsilon}{2} \frac{d\theta(t + \epsilon)}{d t}\\
& = {} \theta(t + \epsilon/2) + \frac{\epsilon}{2} v(t+\epsilon)
\end{align*}$|



# Symplectic Integration [2 minutes]

```python
import matplotlib.pyplot as plt; import numpy as np; t = np.linspace(0,2*np.pi, 100); fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(16,8)) 
for i in range(3):
    ax[i].plot(np.cos(t), -np.sin(t)); ax[i].set_aspect('equal', adjustable='box')
theta_t, v_t, t, epsilon = 0, 1, 0, .3; ax[0].set_title("Euler's Method (not Symplectic)")
for i in range(20):
    v_t_e = v_t - epsilon * theta_t; theta_t_e = theta_t + epsilon * v_t; ax[0].plot([theta_t, theta_t_e], [v_t, v_t_e]) 
    theta_t, v_t, t = theta_t_e, v_t_e, t+epsilon  
theta_t, v_t, t, epsilon = 0, 1, 0, .75; ax[1].set_title("Modified Euler's (First Order Symplectic)")
for i in range(50):
    v_t_e = v_t - epsilon * theta_t; theta_t_e = theta_t + epsilon * v_t_e; ax[1].plot([theta_t, theta_t_e], [v_t, v_t_e]) 
    theta_t, v_t, t = theta_t_e, v_t_e, t+epsilon
theta_t, v_t, t, epsilon = 0, 1, 0, .1; ax[2].set_title("The Leapfrog Method (Second Order Symplectic)")
for i in range(50):
    v_t_e = v_t - epsilon / 2 * theta_t; theta_t_e = theta_t + epsilon * v_t_e; v_t_e = v_t_e - epsilon / 2 * theta_t_e; ax[2].plot([theta_t, theta_t_e], [v_t, v_t_e]) 
    theta_t, v_t, t = theta_t_e, v_t_e, t+epsilon
```

# Return of the ~~Jedi~~ Energy Contour Proposals [5 minutes]

- Errors in ***Euler's method***, even small numerical errors, compound over time... <font style="color:red">quickly failing to satisfy ***conservation of energy***</font>

- ***Symplectic integrators*** like the ***modified Euler's*** and ***leapfrog methods*** correct misteps over time and so maintain the energy level of the ***Hamiltonian dynamics system*** and thus kind of obey ***the conservation of energy law***... **<font style="color:blue">BUT ONLY ON AVERAGE!!!</font>**

    - While ***symplectic integrators*** are theoretically ***time-reversible***, "drift" along with numeric error combine to make "reversing" trajectories back to their initial values unrealisitic... <font style="color:purple">so minimizing "drift" and numeric error important!</font>
    - "Drift" depends on the ***order*** of the ***symplectic integrator*** and the ***time discretization step size*** of the ***integration***
 

$$\begin{array}{c}\text{"Drift" affects acceptance rates}\\\color{gray}{\text{as do energy perturbations }\epsilon}\end{array}\quad \require{AMSmath}\min\Bigg(1,\overset{\neq 1 \text{ from contour drift}}{\boxed{\frac{p(\theta_t, v_t\color{gray}{+\epsilon}|x)}{p(\theta_0, v_0|x)}}}\underset{\text{making symmetry approximate}}{\overset{\neq 1 \text{ drift breaks symmetry}}{\boxed{\frac{\tilde b(\theta_0, v_0|\theta_t, v_t\color{gray}{+\epsilon})}{\tilde b(\theta_t, v_t\color{gray}{+\epsilon}|\theta_0, v_0)}}}}\Bigg)$$

# HMC Diagnostics: Divergences, Autocorrelation<br>(Effective Sample Size), and Acceptance Rates [10 minutes]

- Discrete approximations of ***Hamiltonian trajectories*** should ideally follow ***energy contours*** so $p(\theta_t,v_t|x) \approx p(\theta_0,v_0|x)$
- Accumulating numerical drift is never fully avoidable though, so there are practical limits on ***trajectory*** lengths 
    - In high dimensional space "drift" will inevitably be away from the probability mass which is relatively highly localized
- A ***divergence*** $p(\theta_0,v_0|x) \not = p(\theta_t,v_t|x) \approx 0$ results in extremely low ***Metroplis-Hastings acceptance rates***
    - This is caused by prolongued "drift" or excessive ***curvature*** in the ***energy function (log joint distribution)***
    - If the ***energy function (log joint distribution)*** is changing so rapidly that a step from $(\theta_t,v_t)$ to $(\theta_{t+1},v_{t+1})$<br><font style="color:purple">[based on the ***derivative*** at $(\theta_t,v_t)$ at time $t$]</font> are insufficient to accurately <font style="color:gray">(or at least approximately)</font><br>follow the ***energy countour curvature***, then $(\theta_{t+1},v_{t+1})$ will "fly off" ("step right off") the distributional mass
    - <u>Shorter ***discrete time steps*** can address this, but then equivalent ***trajectory*** evolutions need more computational steps</u>

##  In HMC each <font style="color:gray">(Hamiltonian discretized differential equations)</font> step incurs computational cost; but, you must make these steps in order to move proposals further away far current values; otherwise, autocorrelation is high and effective sample size is small(!)

- HMC is most costly for ***posterior distributions*** with excessive ***curvature*** requiring many very small steps to create proposals
- ***High autocorrelation*** and ***low Metropolis-Hastings acceptance rates*** mean it's computationally expensive to make proposals, and the proposals can't be moved far away and are only being accepted when they're still "close" to the prevous values...
    - In this is happening, the HMC algorithm is failing and inefficient <font style="color:gray">(and the model may be intractable and unusable)</font>
    

# HMC Automatic Tuning and<br>Algorithmic Perforemance Assessment [5 minutes]

An interesting characteristic of ***HMC*** is that its algorithmic performance can be inferred form its runtime behavior

- ***Energy countour trajectory step sizes*** are automatically tuned in ***HMC*** algorithms<br>to optimally achieve an optimal ***Metropolis-Hastings acceptance rate*** (usuall of $80\%$)


  - ***Leapfrog integration*** is a key component of ***HMC*** since creating long ***energy contour trajectories*** with high ***Metropolis-Hastings acceptance rate*** is what reduces ***autocorrelation*** and increases ***effective sample size***
  
$$n_{\text{eff}} = \frac{n}{\sum_{t=-\infty}^\infty\rho_t} = \frac{n}{1+2\sum_{t=1}^\infty\rho_t} \quad \text{ for order $t$ autocorrelations } \quad \rho_t $$
