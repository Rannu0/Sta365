Dataset: https://www.kaggle.com/datasets/dileep070/heart-disease-prediction-using-logistic-regression

```python
import pandas as pd
import pymc as pm

data_path = '/content/framingham.csv'
df = pd.read_csv(data_path)

# Remove missing values
df = df.dropna()

# Select a subset of variables for the analysis
X = df[['male', 'age', 'currentSmoker', 'cigsPerDay', 'BPMeds', 'prevalentStroke', 'prevalentHyp', 'diabetes', 'totChol', 'sysBP', 'diaBP', 'BMI', 'heartRate', 'glucose']]
y = df['TenYearCHD']

# Define the Bayesian Logistic Regression Model
with pm.Model() as logistic_model:
    intercept = pm.Normal('Intercept', mu=0, sigma=10)
    beta = pm.Normal('beta', mu=0, sigma=10, shape=X.shape[1])

    # Prepare data for model
    X_shared = pm.Data('X_shared', X)
    y_shared = pm.Data('y_shared', y)

    # Linear combination of inputs
    mu = intercept + pm.math.dot(X_shared, beta)

    # Logistic regression likelihood
    observed = pm.Bernoulli('obs', logit_p=mu, observed=y_shared)

    # Sample from the posterior
    trace = pm.sample(1000, tune=1000, target_accept=0.95)

summary = pm.summary(trace)
summary
```

## Part II

For a normal (Gaussian) prior on $\beta$, the prior distribution is:
$$ p(\beta) \propto \exp\left(-\frac{1}{2\sigma^2}\sum_{j=1}^p (\beta_j - b_j)^2\right) $$

Given $\sigma=1$ and $b_j=0$, this simplifies to:
$$ p(\beta) \propto \exp\left(-\frac{1}{2}\sum_{j=1}^p \beta_j^2\right) $$

The log posterior, combining this prior with the Gaussian likelihood, is proportional to:
$$ -\frac{1}{2}||y-X\beta||_2^2 - \lambda ||\beta||_2^2 $$

For a Laplace (double exponential) prior on $\beta$, the prior distribution is:
$$ p(\beta) \propto \exp\left(-\frac{\lambda}{\sigma}\sum_{j=1}^p |\beta_j - b_j|\right) $$

Again, with $\sigma=1$ and $b_j=0$, this becomes:
$$ p(\beta) \propto \exp\left(-\lambda\sum_{j=1}^p |\beta_j|\right) $$

The log posterior, combining this prior with the Gaussian likelihood, is proportional to:
$$ -\frac{1}{2}||y-X\beta||_2^2 - \lambda ||\beta||_1 $$
