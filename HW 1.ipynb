### Try One Option and Collect Data:
We will first try the "Red" one and collect data. We assume that $[\theta_{A}, \theta_{B}, \theta_{C}]  = [0.3, 0.5, 0.7]$

```python
from scipy import stats
import numpy as np

# set seed
np.random.seed(1007)

# Define prior parameters for "Red"
alpha_red, beta_red, n_red, x_red = 1, 1, 0, 0
success = 0
for i in range(50):
    if stats.binom(n=1,p=0.3).rvs(size=1)[0] == 1:
        success += 1
success
```

We set the seed as 1007649617 for reproducibility. Next, we defined the prior parameters. Initially, we set alpha=1 and beta=1. This results in a uniform distribution, because we assume all probabilities of winning are equally likely.

This code snippet is using a `for` loop to iterate over the range of numbers from 0 to 9 (inclusive), and for each iteration, it generates a random sample from a binomial distribution with parameters `n=1` and `p=0.5`. The condition `== 1` checks if the generated value is equal to 1 (considered a success in a Bernoulli trial). If the condition is true, `success` is updated by adding 1.

### Update Posterior:

```python
# Update parameters based on simulated data for "Red"
alpha_posterior_red = alpha_red + success - 1
beta_posterior_red = beta_red + 50 - success - 1

# Create a Beta distribution object for "Red"
beta_distribution_red = stats.beta(a=alpha_posterior_red, b=beta_posterior_red)
beta_distribution_red.pdf(0.3)
```

The code adjusts parameters for a Beta distribution linked to the color "Red" using Bayesian updating, considering simulated success and failure data. Subsequently, it creates a new Beta distribution with the updated parameters and calculates the likelihood of observing the value 0.3, which is 2.12. 

### Try The Next Option:

We conducted the similar procedures on the "Blue" one and the "Other" one. Then, we can use a plot to choose the best one.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

np.random.seed(1007)
# Define prior parameters for "Red"
alpha_red, beta_red, n_red, x_red = 1, 1, 0, 0

# Simulate data for "Red" option (theta_red = 0.3)
for i in range(50):
    if stats.binom(n=1, p=0.3).rvs(size=1)[0] == 1:
        x_red += 1    
    n_red += 1

# Update parameters based on simulated data for "Red"
alpha_posterior_red = alpha_red + x_red - 1
beta_posterior_red = beta_red + n_red - x_red - 1

# Create a Beta distribution object for "Red"
beta_distribution_red = stats.beta(a=alpha_posterior_red, b=beta_posterior_red)

# Define prior parameters for "Blue"
alpha_blue, beta_blue, n_blue, x_blue = 1, 1, 0, 0

# Simulate data for "Blue" option (theta_blue = 0.5)
for i in range(50):
    if stats.binom(n=1, p=0.5).rvs(size=1)[0] == 1:
        x_blue += 1    
    n_blue += 1

# Update parameters based on simulated data for "Blue"
alpha_posterior_blue = alpha_blue + x_blue - 1
beta_posterior_blue = beta_blue + n_blue - x_blue - 1

# Create a Beta distribution object for "Blue"
beta_distribution_blue = stats.beta(a=alpha_posterior_blue, b=beta_posterior_blue)

# Define prior parameters for "Other"
alpha_other, beta_other, n_other, x_other = 1, 1, 0, 0

# Simulate data for "Other" option (theta_other = 0.7)
for i in range(50):
    if stats.binom(n=1, p=0.7).rvs(size=1)[0] == 1:
        x_other += 1    
    n_other += 1

# Update parameters based on simulated data for "Other"
alpha_posterior_other = alpha_other + x_other - 1
beta_posterior_other = beta_other + n_other - x_other - 1

# Create a Beta distribution object for "Other"
beta_distribution_other = stats.beta(a=alpha_posterior_other, b=beta_posterior_other)

# Plot the Beta distributions for "Red", "Blue", and "Other"
x_values = np.linspace(0, 1, 1000)
pdf_values_red = beta_distribution_red.pdf(x_values)
pdf_values_blue = beta_distribution_blue.pdf(x_values)
pdf_values_other = beta_distribution_other.pdf(x_values)

plt.plot(x_values, pdf_values_red, label='Red', color ="red")
plt.plot(x_values, pdf_values_blue, label='Blue', color = "blue")
plt.plot(x_values, pdf_values_other, label='Other', color = "orange")

plt.title('Posterior Beta Distributions for "Red", "Blue", and "Other" Options')
plt.xlabel('Probability of Success')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

In the plot, we can see that the "Other" one is more narrow, we choose the "Other" one.
